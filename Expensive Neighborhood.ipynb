```python
manhattan2020 = pd.read_csv('2020/2020_manhattan.csv')
manhattan2019 = pd.read_csv('2019/2019_manhattan.csv')
manhattan2018 = pd.read_csv('2018/2018_manhattan.csv')
manhattan2017 = pd.read_csv('2017/2017_manhattan.csv')
manhattan2016 = pd.read_csv('2016/2016_manhattan.csv')
```

```python
bronx2020 = pd.read_csv('2020/2020_bronx.csv')
bronx2019 = pd.read_csv('2019/2019_bronx.csv')
bronx2018 = pd.read_csv('2018/2018_bronx.csv')
bronx2017 = pd.read_csv('2017/2017_bronx.csv')
bronx2016 = pd.read_csv('2016/2016_bronx.csv')
```

```python
queens2020 = pd.read_csv('2020/2020_queens.csv')
queens2019 = pd.read_csv('2019/2019_queens.csv')
queens2018 = pd.read_csv('2018/2018_queens.csv')
queens2017 = pd.read_csv('2017/2017_queens.csv')
queens2016 = pd.read_csv('2016/2016_queens.csv')
```

```python
staten_island2020 = pd.read_csv('2020/2020_staten_island.csv')
staten_island2019 = pd.read_csv('2019/2019_statenisland.csv')
staten_island2018 = pd.read_csv('2018/2018_statenisland.csv')
staten_island2017 = pd.read_csv('2017/2017_statenisland.csv')
staten_island2016 = pd.read_csv('2016/2016_statenisland.csv')
```

```python
manhattanframes = [manhattan2016,manhattan2017,manhattan2018,manhattan2019,manhattan2020]
brooklynframes = [brooklyn2016,brooklyn2017,brooklyn2018,brooklyn2019,brooklyn2020]
bronxframes = [bronx2016,bronx2017,bronx2018,bronx2019,bronx2020]
queensframes = [queens2016, queens2017,queens2018,queens2019,queens2020]
statenislandframes = [staten_island2016,staten_island2017,staten_island2018,staten_island2019,staten_island2020]
```

```python
manhattan2020.columns = ['BOROUGH', 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY',
       'TAX CLASS AT PRESENT', 'BLOCK', 'LOT', 'EASE-MENT',
       'BUILDING CLASS AT PRESENT', 'ADDRESS', 'APARTMENT NUMBER',
       'ZIP CODE', 'RESIDENTIAL UNITS', 'COMMERCIAL UNITS',
       'TOTAL UNITS', 'LAND SQUARE FEET', 'GROSS SQUARE FEET',
       'YEAR BUILT', 'TAX CLASS AT TIME OF SALE',
       'BUILDING CLASS AT TIME OF SALE', 'SALE PRICE', 'SALE DATE']
```

```python
manhattan2016.columns = manhattan2020.columns
manhattan2017.columns = manhattan2020.columns
manhattan2018.columns = manhattan2020.columns
manhattan2019.columns = manhattan2020.columns
```

```python
brooklyn2016.columns = manhattan2020.columns
brooklyn2017.columns = manhattan2020.columns
brooklyn2018.columns = manhattan2020.columns
brooklyn2019.columns = manhattan2020.columns
brooklyn2020.columns = manhattan2020.columns
```

```python
bronx2016.columns = manhattan2020.columns
bronx2017.columns = manhattan2020.columns
bronx2018.columns = manhattan2020.columns
bronx2019.columns = manhattan2020.columns
bronx2020.columns = manhattan2020.columns
```

```python
queens2016.columns = manhattan2020.columns
queens2017.columns = manhattan2020.columns
queens2018.columns = manhattan2020.columns
queens2019.columns = manhattan2020.columns
queens2020.columns = manhattan2020.columns
```

```python
staten_island2016.columns = manhattan2020.columns
staten_island2017.columns = manhattan2020.columns
staten_island2018.columns = manhattan2020.columns
staten_island2019.columns = manhattan2020.columns
staten_island2020.columns = manhattan2020.columns
```

```python
manhattanmerge = pd.concat(manhattanframes,ignore_index=True)
```

```python
brooklynmerge = pd.concat(brooklynframes,ignore_index=True)
```

```python
bronxmerge = pd.concat(bronxframes,ignore_index=True)
```

```python
queensmerge = pd.concat(queensframes,ignore_index=True)
```

```python
statenislandmerge = pd.concat(statenislandframes,ignore_index=True)
```

```python
manhattanmerge = manhattanmerge.drop(['BOROUGH','BUILDING CLASS CATEGORY',
       'TAX CLASS AT PRESENT', 'BLOCK', 'LOT', 'EASE-MENT',
       'BUILDING CLASS AT PRESENT', 'ADDRESS', 'APARTMENT NUMBER',
       'ZIP CODE', 'RESIDENTIAL UNITS', 'COMMERCIAL UNITS',
       'TOTAL UNITS', 'LAND SQUARE FEET', 'GROSS SQUARE FEET',
       'YEAR BUILT', 'TAX CLASS AT TIME OF SALE',
       'BUILDING CLASS AT TIME OF SALE','SALE DATE'], axis = 1)
```

```python
manhattanmerge.columns
```

```python
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].astype(str)
```

```python
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].apply(lambda x: x.replace(',', ''))
```

```python
manhattanmerge['SALE PRICE'].unique()
```

```python
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].replace(' -   ', '0')
```

```python
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].replace('$0', '0')
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].replace('$4844809', '4844809')
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].replace(' ', '0')
```

```python
manhattanmerge['SALE PRICE'] = pd.to_numeric(manhattanmerge['SALE PRICE'], errors = 'coerce')
```

```python
manhattanmerge['SALE PRICE'].isnull().sum()
```

```python
manhattanmerge.dropna(inplace = True)
```

```python
manhattanmerge.count()
```

```python
manhattanmerge.isna().sum()
```

```python
manhattanmerge['SALE PRICE'].sum()/manhattanmerge['SALE PRICE'].count()
```

```python
manhattanmerge['SALE PRICE'] = manhattanmerge['SALE PRICE'].astype(float).div(1000000)
df = manhattanmerge.groupby(['NEIGHBORHOOD']).sum().reset_index()
```

```python
plt.figure(figsize=(20,6))
plt.title('Manhattan Prices Comparision (2016-2020)')
plt.xlabel('Neighbourhood')
plt.ylabel('Prices in Millions')
plt.bar(df['NEIGHBORHOOD'],df['SALE PRICE'], color = (0.5,0.1,0.5,0.6))
plt.xticks(rotation=90)
plt.savefig('Manhattan Prices Comparision (2016-2020)',bbox_inches='tight')
plt.show()
```
